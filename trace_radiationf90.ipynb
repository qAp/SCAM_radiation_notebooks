### cloud_sw()

**Input:**

*(ncol, nlay)*
* fraction (fcld)
* snow fraction (fcldsnow)
* ice water path (iwp)
* liquid water path (lwp)
* snow water path (swp)
* effective radius of cloud ice (reff_ice)
* effective radius of cloud liquid (reff_liq)
* effective radius of snow (reff_snw)

*boolean*
* option for scaling cloud optical thickness (overcast)

**Output:**

*(ncol, nlay, nuvband)*
* optical thickness (tauc_uv)
* optical thickness for direct radiation (tauc_bm_uv)
* optical thickness for diffuse radiation (tauc_df_uv)
* asymmetry factor (asyc_uv)

*(ncol, nlay, nirband)*
* optical thickness (tauc_ir)
* optical thickness for direct radiation (tauc_bm_ir)
* optical thickness for diffuse radiation (tauc_df_ir)
* asymmetry factor (asyc_ir)
* single scattering albedo (ssac_ir)

*(ncol, 3)*

cloud cover of high, middle, low clouds (cc)

*(ncol)*
* level index separated high, middle cloud group, about 400 hPa (idx_ct)
* level index separated middle, low cloud group, about 700 hPa (idx_cb)
* level index of highest cloud top (idx_cldtop)
