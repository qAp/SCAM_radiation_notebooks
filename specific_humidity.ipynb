```python
%matplotlib inline
import os
import itertools
import numpy as np
import pandas as pd
import netCDF4
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as matcm
```

```python
filepath_rrtmg = '../examples/yihsuans_arm95_rrtmg/0209-123711/camrun.cam.h0.0095-07-18-19800.nc'
filepath_clirad = '../examples/yihsuans_arm95_clirad/0210-133728/camrun.cam.h0.0095-07-18-19800.nc'
```

```python
# load rrtmg results

h0rrtmg = netCDF4.Dataset(filepath_rrtmg, 'r')

time_rrtmg = h0rrtmg.variables['time']
lev_rrtmg = h0rrtmg.variables['lev']
q_rrtmg = h0rrtmg.variables['Q']

h0rrtmg.close()
```

```python
print(q_rrtmg)
```

```python
q_rrtmg_data = q_rrtmg[:, :, 0, 0]
q_rrtmg_units = q_rrtmg.units
q_rrtmg_long_name = q_rrtmg.long_name

time_rrtmg_data = time_rrtmg[:]
time_rrtmg_units = time_rrtmg.units
time_rrtmg_long_name = time_rrtmg.long_name

lev_rrtmg_data = lev_rrtmg[:]
lev_rrtmg_units = lev_rrtmg.units
lev_rrtmg_long_name = lev_rrtmg.long_name

X, Y = np.meshgrid(time_rrtmg_data, lev_rrtmg_data)
Z = q_rrtmg_data.T

matplotlib.rcParams.update({'font.size': 13})
fig = plt.figure(figsize = (8, 7))
ax = fig.add_subplot(111)
cs = ax.contourf(X, Y, Z,
                 cmap = matcm.coolwarm,
                 extend = 'both',
                 levels = np.arange(0., 0.025, 0.001))

cbar = plt.colorbar(cs, ax = ax)
cbar.ax.set_ylabel('{}'.format(q_rrtmg_units))

ax.set_xlabel('{} [{}]'.format(time_rrtmg_long_name, time_rrtmg_units))
ax.set_ylabel('{} [{}]'.format(lev_rrtmg_long_name, lev_rrtmg_units))
ax.set_title('{}'.format(q_rrtmg_long_name))

ax.grid(b = True)
ax.invert_yaxis()
```
