```python
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
def get_casenames():
    pass

def filepaths_spcvmc_cldfrac(mcica = True, cldfprime0 = False):
    if mcica and not cldfprime0:
        return '../examples/MPACE/cam5_3hrly_LWrrtmg_SWrrtmg/sw_spcvmc_pcldfmc.dat'
    
    if mcica and cldfprime0:
        return '../examples/MPACE/cam5_3hrly_LWrrtmg_SWrrtmg_cldfprime0/sw_spcvmc_pcldfmc.dat'
    
    if not mcica and cldfprime0:
        return '../examples/MPACE/cam5_3hrly_LWrrtmg_SWrrtmg_SWnoMcICA_cldfprime0/sw_spcvmc_pcldfmc.dat'
    
    if not mcica and not cldfprime0:
        return '../examples/MPACE/cam5_3hrly_LWrrtmg_SWrrtmg_SWnoMcICA/sw_spcvmc_pcldfmc.dat'
    
    
```

```python
def get_cldfrac_overlaps(df):
    '''df should be from read_csv of a No McICA run, since only these runs have different 
    overlaps, combinations of overlaps.
    '''
    n_step, g_point = 0, 1
    
    dfmcica_step0 = pd.read_csv(filepaths_spcvmc_cldfrac(), 
                                header = 0, sep = r'\s+', index_col = [0, 1, 2])\
                                .loc[(n_step, slice(None), slice(None))]
    return pd.concat([df.iloc[(overlap - 1) * len(dfmcica_step0): overlap * len(dfmcica_step0)]\
                      .loc[(n_step, slice(None), g_point)] for overlap in range(1, 8 + 1)],
                      axis = 1,
                      keys = range(1, 8 + 1))        
    
```

```python
df_nomcica = pd.read_csv(filepaths_spcvmc_cldfrac(mcica = False),
                        header = 0, sep = r'\s+', index_col = [0, 1, 2])
get_cldfrac_overlaps(df_nomcica)
```
