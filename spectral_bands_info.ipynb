```python
%matplotlib inline
import os
import sys
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt

import rtmtools.clirad.sw.info as cliradswinfo
import rtmtools.clirad.lw.info as cliradlwinfo
import rtmtools.rrtmg.sw.info as rrtmgswinfo
import rtmtools.rrtmg.lw.info as rrtmglwinfo

from IPython.display import HTML
```

```python
def plt_band_and_label(ax, lbnd = 0, hbnd = 500, band_number = 5):
    ax.axvline(x = lbnd)
    ax.axvline(x = hbnd)
    ax.text(x = .5 * (lbnd + hbnd), y = .5, 
            s = '{}'.format(band_number), fontsize = 10, color = 'k',
           verticalalignment = 'center', horizontalalignment = 'center')
   

def plot_spectral_bands_sw():
    bands_cliradsw = cliradswinfo.wavenumber_bands()
    bands_rrtmgsw  = rrtmgswinfo.wavenumber_bands()

    fig, axs = plt.subplots(nrows = 2, ncols = 1, figsize = (15, 6))

    for band_number, bands in bands_cliradsw.items():
        for lbnd, hbnd in bands:
            plt_band_and_label(axs[0], lbnd = lbnd, hbnd = hbnd, band_number = band_number)

    axs[0].yaxis.set_ticklabels([])
    axs[0].set_ylabel('CLIRAD-SW', fontsize = 15.)
        
    for band_number, bands in bands_rrtmgsw.items():
        for lbnd, hbnd in bands:
            plt_band_and_label(axs[1], lbnd = lbnd, hbnd = hbnd, band_number = band_number)
        
    axs[1].set_xlim((0, 60000))
    axs[1].set_xlabel('wavenumber [$cm^{-1}$]', fontsize = 15)

    axs[1].yaxis.set_ticklabels([])
    axs[1].set_ylabel('RRTMG-SW', fontsize = 15)

    fig.suptitle('Shortwave spectral bands', fontsize = 16)
    
    
def plot_spectral_bands_lw():
    bands_cliradlw = cliradlwinfo.wavenumber_bands()
    bands_rrtmglw  = rrtmglwinfo.wavenumber_bands()

    fig, axs = plt.subplots(nrows = 2, ncols = 1, figsize = (15, 6))

    for band_number, bands in bands_cliradlw.items():
        for lbnd, hbnd in bands:
            plt_band_and_label(axs[0], lbnd = lbnd, hbnd = hbnd, band_number = band_number)

    axs[0].set_xlim((0, 3500))
    axs[0].yaxis.set_ticklabels([])
    axs[0].set_ylabel('CLIRAD-LW', fontsize = 15.)
        
    for band_number, bands in bands_rrtmglw.items():
        for lbnd, hbnd in bands:
            plt_band_and_label(axs[1], lbnd = lbnd, hbnd = hbnd, band_number = band_number)
        
    axs[1].set_xlim((0, 3500))
    axs[1].set_xlabel('wavenumber [$cm^{-1}$]', fontsize = 15)

    axs[1].yaxis.set_ticklabels([])
    axs[1].set_ylabel('RRTMG-LW', fontsize = 15)

    fig.suptitle('Longwave spectral bands', fontsize = 16)
```

# Spectral bands in RRTMG and CLIRAD

```python
plot_spectral_bands_sw()
plot_spectral_bands_lw()
```
