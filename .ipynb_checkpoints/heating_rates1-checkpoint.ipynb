```python
%matplotlib inline
import numpy as np
import pandas as pd
import xray
import itertools

import matplotlib.pyplot as plt
```

```python
filepath_rrtmg = '../examples/yihsuans_arm95_rrtmg/0209-123711/camrun.cam.h0.0095-07-18-19800.nc'
```

```python
h0rrtmg = xray.open_dataset(filepath_rrtmg, decode_times = False)
```

#### convert units of lev

```python
def pres_hybrid_ccm(ps, p0, hyam, hybm):
    '''
    Calculates pressure (hPa/mbar) at hybrid levels 
    (https://www.ncl.ucar.edu/Document/Functions/Built-in/pres_hybrid_ccm.shtml)
    INPUT:
    ps --- [Pa]
    p0 --- [Pa]
    OUTPUT:
    pressure at hybrid levels [mbar]
    '''
    return 1e-2 * (hyam * p0 + hybm * ps).transpose('time', 'lev', 'lat', 'lon')
```

```python
pressure = pres_hybrid_ccm(h0rrtmg['PS'],
                           h0rrtmg['P0'],
                           h0rrtmg['hyam'],
                           h0rrtmg['hybm'])
```

#### convert time to time stamps    

```python
h0rrtmg['time'].attrs['units']
```

```python
t0 = pd.datetime(1995, 7, 18, 5, 30, 0)
timestamps = [t0 + pd.DateOffset(days = doffset) for doffset in h0rrtmg['time'].values]
```

#### convert units of qrl from ${}^{\circ} s^{-1}$ to ${}^{\circ} day^{-1}$

```python
def degpersec2degperday(qrl):
    qrl *= 86400.
    qrl.attrs['units'] = 'K/day'
    return qrl
```

#### Copy h0rrtmg['QRL'] and change its coordinates and units

```python
qrl = h0rrtmg['QRL'].copy()

qrl = degpersec2degperday(qrl)

qrl.coords['time'] = ('time', timestamps, {'units': 'datetime'})

qrl.coords['lev'] = ('lev', pressure[{'time': 0, 'lon': 0, 'lat': 0}],
                     {'units': 'mbar'})

qrl.coords['time'].attrs['units']
```

#### pandas-plot qrl vs time for the first 3 levs
Notice how the time labels are formatted

```python
qrl.squeeze().to_pandas().iloc[:, : 3].plot(figsize = (7, 7))
```

#### contour-fill plot time vs pressure vs qrl

```python
import matplotlib.dates as mdates
#X, Y = np.meshgrid(dts, qrl['lev'].values)

Z = qrl.values[:, :, 0, 0].T
x = np.array([pd.Timestamp(t).to_datetime() for t in qrl['time'].values])
y = qrl['lev'].values
```

```python
# this plot method only works if datetimes are converted to
# numbers first
X = np.tile(x.reshape((1, x.shape[0])), (y.shape[0], 1))
Y = np.tile(y.reshape((y.shape[0], 1)), (1, x.shape[0]))

X = matplotlib.dates.date2num(X)

plt.contourf(X, Y, Z)
```

```python
# this plot method does not work
# numpy.meshgrid does not take datetime objects
X, Y = np.meshgrid(x, y)
```

```python
# this method works
# plt.contourf() can acutally take as vectors the X and Y variables
# so there is no need for numpy.tile or numpy.meshgrid

import matplotlib.cm as matcm

Z = qrl.values[:, :, 0, 0].T
x = np.array([pd.Timestamp(t).to_datetime() for t in qrl['time'].values])
y = qrl['lev'].values

fig = plt.figure(figsize = (7, 6), dpi = 500)

ax = fig.add_subplot(111)

contourf_levels = np.arange(-5, 5, .5)

cs = ax.contourf(x, y, Z,
                 levels = contourf_levels, 
                 cmap = matcm.coolwarm,
                 extend = 'both')

# colour map
cs.cmap.set_under(color = 'black')
cs.cmap.set_over(color = 'yellow')

# colour bar
cbar = plt.colorbar(cs, ax = ax)
cbar.ax.set_ylabel('K/day')

# title
ax.set_title('heating rate')

# y-axis
ax.invert_yaxis()
ax.set_ylabel('pressure [mbar]')
ax.locator_params(axis = 'y', nbins = 10)
ax.yaxis.grid(b = True, which = 'major')

# x-axis
ax.set_xlabel('time [datetime]')
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('\n%b\n%Y'))
ax.xaxis.set_minor_locator(mdates.DayLocator(interval = 2))
ax.xaxis.set_minor_formatter(mdates.DateFormatter('%d'))
ax.xaxis.grid(b = True, which = 'minor')

#fig.autofmt_xdate()
```
