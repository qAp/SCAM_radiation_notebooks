## ARM95 | LW CLIRAD | SW RRTMG
This version of CLIRAD contains Yi Hsuan's min/max statements in the
subroutine cloud_lw().  These statements do the following to control the value of computed cloud optical thickness with scattering:

* if the computed cloud optical thickness with scattering is less than zero, then it is set to zero
* if the computed cloud optical thickness with scatterring is greater than that without scattering, it is set to the optical thickness *with no* scattering
* if the computed cloud optical thickness with scattering is greated than zero and less than that without scattering, it is left alone.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import xray
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams.update({'font.size': 16})
```

```python
filepath = '../examples/arm95_LWclirad_SWrrtmg/0306-111654_yihsuan_minmax_scattering/camrun.cam.h0.0095-07-18-19800.nc'
```

```python
h0 = xray.open_dataset(filepath, decode_times = False)
```

#### convert units of lev

\begin{equation}
(\mathrm{hyam} * \mathrm{P0} + \mathrm{hybm} * \mathrm{PS}) / 100 
\end{equation}

gives pressure in mbars.

```python
pressure = 1e-2 * (h0['hyam'] * h0['P0'] + h0['hybm'] * h0['PS'])\
.transpose('time', 'lev', 'lat', 'lon')
pressure.attrs['units'] = 'mbar'
pressure = pressure[{'time': 0, 'lat': 0, 'lon': 0}]
```

#### convert time to datetime objects

```python
t0 = pd.to_datetime(h0['time'].units.split('since')[-1]) 
timestamps = [t0 + pd.DateOffset(days = dayssince) \
              for dayssince in h0['time'].values]
datetimes = np.array([stamp.to_datetime() for stamp in timestamps])
```

### Cloud fraction

#### make a copy of CLOUD (cloud fraction) and change its lev and time coordindate to the new ones

```python
cloud = h0['CLOUD'].copy()
cloud.coords['time'] = ('time', datetimes.copy(), {'units': 'datetime'})
cloud.coords['lev'] = ('lev', pressure.copy(), {'units': 'mbar'})
```

#### plot: contourf of time vs pressure vs CLOUD

```python
x = [pd.Timestamp(dtns).to_datetime() \
     for dtns in cloud.coords['time'].values]
y = cloud.coords['lev'].values
Z = cloud[{'lon': 0, 'lat': 0}].transpose('lev', 'time').values


fig = plt.figure(figsize = (7, 6))
ax = fig.add_subplot(111)

cs = ax.contourf(x, y, Z,
                 cmap = matplotlib.cm.coolwarm,
                 levels = np.arange(Z.min(), Z.max() + .05, .05))

cs.cmap.set_over('yellow')
cs.cmap.set_under('black')

cbar = plt.colorbar(cs)
cbar.set_label('[{}]'.format(cloud.units))

ax.set_title('{}'.format(cloud.attrs['long_name']))

# x-axis
ax.set_xlabel('time [{}]'.format(cloud['time'].attrs['units']))
ax.xaxis.set_major_locator(matplotlib.dates.MonthLocator())
ax.xaxis.set_major_formatter(matplotlib.dates.DateFormatter('\n%b\n%Y'))
ax.xaxis.set_minor_locator(matplotlib.dates.DayLocator(interval = 2))
ax.xaxis.set_minor_formatter(matplotlib.dates.DateFormatter('%d'))
ax.xaxis.grid(which = 'minor')

# y-axis
ax.set_ylabel('lev [{}]'.format(cloud['lev'].attrs['units']))
ax.invert_yaxis()
ax.yaxis.grid(which = 'major')
```

### LW heating rate

#### make a copy of QRL (longwave heating rate), change its units to deg/day, its lev and time coordinates to pressure and datetime

```python
qrl = getattr(h0, 'QRL', None).copy()
qrl.coords['time'] = ('time', datetimes.copy(), {'units': 'datetime'})
qrl.coords['lev'] = ('lev', pressure.copy(), {'units': pressure.units})
qrl.values *= 86400.
qrl.units = 'K/day'
```

#### plot: contourf of time vs pressure vs LW heating rate 

```python
x = [pd.Timestamp(dtns).to_datetime() \
     for dtns in qrl.coords['time'].values]
y = qrl.coords['lev'].values
Z = qrl[{'lon': 0, 'lat': 0}].transpose('lev', 'time').values

fig = plt.figure(figsize = (7, 6))
ax = fig.add_subplot(111)


cs = ax.contourf(x, y, Z,
                 cmap = matplotlib.cm.coolwarm,
                 levels = np.arange(-5, 5 + .5, .5),
                 extend = 'both')

cbar = plt.colorbar(cs)
cbar.set_label('[{}]'.format(qrl.units))

ax.set_title('{}'.format(qrl.attrs['long_name']))

# yaxis
ax.set_ylabel('lev [{}]'.format(qrl['lev'].attrs['units']))
ax.invert_yaxis()
ax.yaxis.grid(which = 'major')

# xaxis
ax.set_xlabel('time [{}]'.format(qrl['time'].attrs['units']))
ax.xaxis.set_major_locator(matplotlib.dates.MonthLocator())
ax.xaxis.set_major_formatter(matplotlib.dates.DateFormatter('\n%b\n%Y'))
ax.xaxis.set_minor_locator(matplotlib.dates.DayLocator(interval = 2))
ax.xaxis.set_minor_formatter(matplotlib.dates.DateFormatter('%d'))
ax.xaxis.grid(which = 'minor')
```

### Specific Humidity

#### make a copy of Q (specific humidity), change its lev and time coordinates to pressure and datetime

```python
q = getattr(h0, 'Q', None).copy()
q.coords['lev'] = ('lev', pressure.copy(), {'units': pressure.units})
q.coords['time'] = ('time', datetimes.copy(), {'units': 'datetime'})
```

#### plot: contourf of time vs pressure vs specific humidity

```python
x = np.array([pd.Timestamp(dtns).to_datetime() \
              for dtns in q.coords['time'].values])
y = q.coords['lev'].values
Z = q[{'lon': 0, 'lat': 0}].transpose('lev', 'time').values

fig = plt.figure(figsize = (7, 6))
ax = fig.add_subplot(111)
cs = ax.contourf(x, y, Z,
                 levels = np.arange(Z.min(), Z.max() + .001, .001),
                 cmap = matplotlib.cm.coolwarm)

# colorbar
cbar_pow = -2
def colorbar_fmt(value, pos):
    return r'{:.2f}'.format(10**(- cbar_pow) * value)

cbar = plt.colorbar(cs,
                    format = matplotlib.ticker.\
                    FuncFormatter(colorbar_fmt))

cbar.set_label(r'$\times 10^{{{}}}$ [{}]'.format(cbar_pow, q.units))



ax.set_title('{}'.format(q.attrs['long_name']))

# y-axis 
ax.set_ylabel('lev [{}]'.format(q['lev'].units))
ax.invert_yaxis()
ax.yaxis.grid(which = 'major')

# x-axis
ax.set_xlabel('time [{}]'.format(q['time'].units))
ax.xaxis.set_major_locator(matplotlib.dates.MonthLocator())
ax.xaxis.set_major_formatter(matplotlib.dates.DateFormatter('\n%b\n%Y'))
ax.xaxis.set_minor_locator(matplotlib.dates.DayLocator(interval = 2))
ax.xaxis.set_minor_formatter(matplotlib.dates.DateFormatter('%d'))
ax.xaxis.grid(which = 'minor')
```
