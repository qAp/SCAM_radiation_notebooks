```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

import re
import numpy as np
import pandas as pd
import sympy
import sympy.abc as syc
sympy.init_printing()

matplotlib.rcParams.update({'font.size': 14})
```

## Definitions

```python
r = sympy.Symbol('r')
W = sympy.Symbol('W')
g = sympy.Symbol('g')

cbeta = sympy.Matrix(sympy.symbols('c_b:4'))
comega = sympy.Matrix(sympy.symbols('c_w:4'))
cg = sympy.Matrix(sympy.symbols('c_g:4'))

rpowers = sympy.Matrix([r**i for i in range(len(cbeta))])

def opthick_polynomial(state = 'i'):
    if state == 'i':
        return (cbeta[0] + cbeta[1] / r**cbeta[2])
    else:
        return cbeta.dot(rpowers)
    
def ssa_polynomial():
    
    return comega.dot(rpowers)

def asy_polynomial():
    return cg.dot(rpowers)

def symbol_polynomial(state = 'i', optical_property = 'b'):
    if optical_property == 'b':
        if state == 'i':
            return (cbeta[0] + cbeta[1] / r**cbeta[2])
        else:
            return cbeta.dot(rpowers)
    elif optical_property == 'w':
        return comega.dot(rpowers)
    elif optical_property == 'g':
        return cg.dot(rpowers)
```

## Get expansion coefficients from radiation.f90

```python
def regex_numbers():
    return re.compile(r'''

    -?
    \d
    \.
    \d+
    (?: e-?\d)?

    ''', re.VERBOSE)


def regex_row_string(state = 'i', optical_property = 'b', order = 1):
    '''
    INPUT:
    state --- 'i' for ice, 'w' fr liquid
    optical_property --- 'b' for optical thickness
                         'w' for single scattering albedo
                         'g' for asymmetry factor
    '''
    varname = 'a' + state + optical_property
    regex = r'{varname}\({order},:\) = \(.*\n.*\)'\
            .format(varname = varname, order = order)
    return re.compile(regex)


def get_optical_property_coefficients(textstring, state = 'i', 
                                      optical_property = 'b'):
    if optical_property == 'b' and state == 'i':
        orders = range(1, 4)
    else:
        orders = range(1, 5)
        
    regexs = [regex_row_string(state = state, 
                               optical_property = optical_property,
                               order = order)\
              for order in orders]
    
#    print(regex_numbers().findall(regexs[0].findall(textstring)))
    coefficients = np.array([regex_numbers().findall(
                regex.findall(textstring)[0])\
                for regex in regexs]).astype('float')
    df = pd.DataFrame(coefficients,
                      index = range(coefficients.shape[0]),
                      columns = range(1, coefficients.shape[1] + 1))
    df.columns.name = 'wavenumber band'
    df.index.name = 'polynomial order'
    return df.T
    


def get_cloud_lw_body():
    fpath = '/nuwa_cluster/home/jackyu/climate_models/cesm1_2_2/models/atm/cam/src/physics/rrtmg/radiation.F90'

    with open(fpath, mode = 'r', encoding = 'utf-8') as file:
        content = file.read()
    
    return content.split('subroutine cloud_lw')[1]
```

```python
cloudlw_body = get_cloud_lw_body()
```

## Get polynomials and roots

```python
def get_poly_funcs_exprs(optical_property = 'b', state = 'w',
                              funcs_only = False):

    coeffs = get_optical_property_coefficients(cloudlw_body, state = state, 
                                     optical_property = optical_property)

    poly_expr = symbol_polynomial(state = state, 
                                  optical_property = optical_property)

    if optical_property == 'b':
        subs_symbol = cbeta
    elif optical_property == 'w':
        subs_symbol = comega
    elif optical_property == 'g':
        subs_symbol = cg
      
    poly_exprs = [poly_expr.subs(list(zip(subs_symbol, coeffs_wb)))\
                 for i, coeffs_wb in coeffs.iterrows()]
    
    deriv_exprs = [sympy.diff(expr, r) for expr in poly_exprs]

    poly_funcs = [sympy.lambdify([r], poly_expr, 'numpy')\
                 for poly_expr in poly_exprs]

    if optical_property == 'b' and state == 'i':
        min_root_finder, max_root_finder = 1, 10000
    else:
        min_root_finder, max_root_finder = - 400, 400
    
    sympy.mpmath.mp.dps = 15
    root_wbs = [sympy.mpmath.findroot(lambda x: poly_func(x), 
                                      [min_root_finder, max_root_finder],
                                     solver = 'pegasus')\
                for poly_func in poly_funcs]
    
    derivs_root = [sympy.N(deriv_expr.subs(r, root))\
                   for root, deriv_expr in zip(root_wbs, deriv_exprs)]

    return poly_expr, poly_exprs, poly_funcs, root_wbs, derivs_root
```

## Plotting function

```python
def plot_poly_funcs_roots(poly_expr, poly_funcs, real_roots,
                         rmin = 0, rmax = 500,
                         textpos = 'up'):
    
    reff = np.linspace(rmin, rmax, 100)

    fig, axs = plt.subplots(nrows = 4, ncols = 3, figsize = (11, 11),
                           sharex = True, sharey = True)

    for wbidx, (ax, func, root) \
    in enumerate(zip(axs.ravel(), poly_funcs, real_roots)):
        ax.plot(reff, func(reff),
               linestyle = '-', linewidth = 1.5)
        ax.axhline(y = 0, color = 'r', alpha = 0.5)
        ax.axvline(x= 0, color = 'r', alpha = 0.5)
        


        
        xlim = ax.get_xlim()
        xrange = xlim[1] - xlim[0]
        ax.set_xlim((xlim[0] - .015 * xrange, xlim[1] + .015 * xrange))
    
        ax.text(x = .5, y = .2 if textpos == 'down' else .8, 
                s = ('band {}:  '.format(wbidx + 1) + \
                     ('{:.3f}'.format(float(root)) if root else '')),
               transform = ax.transAxes,
               verticalalignment = 'center', horizontalalignment = 'center')

    ax.set_ylim((0, 1))
    yticklocs = ax.yaxis.get_majorticklocs()
    interval_ytick = yticklocs[1] - yticklocs[0]       
    ax.set_ylim((yticklocs[0] - .2 * interval_ytick, 
                 yticklocs[-1] + .2 * interval_ytick))


    plt.figtext(x = 0.02, y = .5, s = '${}$'.format(sympy.latex(poly_expr)), 
                rotation = 90, fontsize = 15,
               horizontalalignment = 'center', verticalalignment = 'center')
    plt.figtext(x = .5, y = 0.08, s = '$r_{eff}$', fontsize = 15)
    plt.subplots_adjust(hspace = 0, wspace = 0)    
```

# $r_{eff}$ used in some SCAM run

```python
def get_SCAM_reff():
    fpath = '/nuwa_cluster/home/jackyu/climate_models/scam/examples/arm95_LWclirad_SWrrtmg/test/reff.dat'
    df = pd.read_csv(fpath, sep = r'\s+', header = 0, index_col = [0, 1, 2])

    df.index = df.index.droplevel(level = 1)
    df.index.names = ['timestep', 'layer']

    df = df.sortlevel(1).sortlevel(0)
    return df

def histogram_reff(ax, df_reff, state = 'liq', bins = 40, logy = True):
    df_reff['reff_' + state]\
    .plot(ax = ax, kind = 'hist', 
          color = (.4, .4, .4), alpha = 0.3, linewidth = 0.,
          bins = bins, logy = logy, normed = True)
    ax.set_xlabel('reff_' + state)
    ax.set_ylabel('area normalised frequency of $r_{eff}$')
    return ax

def overlay_max_reffs(ax, roots, derivs_root, logy = True):
    xlim, ylim = ax.get_xlim(), ax.get_ylim()
    xrange, yrange = xlim[1] - xlim[0], ylim[1] - ylim[0]
    
    if logy:
        text_ylocs = np.logspace(np.log10(ylim[0]), 
                                 np.log10(ylim[1]), 10)[:: -1]
    else:
        text_ylocs = np.linspace(ylim[0], ylim[1], 10)[:: -1]
    
    for wb, (root, deriv) in enumerate(zip(roots, derivs_root)):
        if 0 < root < ax.get_xlim()[-1]:
            if deriv < 0:
                limit_type, colour = 'max', (0.792, 0.313, 0.247)
            elif deriv > 0:
                limit_type, colour = 'min', (0.376, 0.168, 0.831)
        
            colour = np.random.rand(3)
            ax.axvline(x = root, color = colour, alpha = .7, linewidth = 1.4)
            ax.text(x = root, 
                    y = text_ylocs[wb],
                    s = ' {}: {}'.format(wb + 1, limit_type),
                    rotation = -90, color = colour, 
                    alpha = 1, size = 12, weight = 'semibold')
    return ax


def reff_fitrange(state = 'liq'):
    '''
    minimum and maximum reff for liquid and ice spheres 
    (according to Prof. Chou's code for computing reff from temperature)
    '''
    if state == 'liq':
        return 4, 20
    elif state in ['ice', 'snw']:
        return 10, 70
    else:
        raise ValueError('state must be either liq, ice or snw')
        
        
def overlay_reff_fitrange(ax, state = 'liq'):
    line_params = zip(['min', 'max'],
                      reff_fitrange(state = state),
                     ['blue', 'red'])
    for nature, value, color in line_params:
        ax.axvline(x = value, color = color, alpha = .7, linewidth = 1.4)
        ax.text(x = value, y = .8 * ax.get_ylim()[-1], 
                s = '{}: {}'.format(nature, value),
                rotation = -90, color = color, alpha = 1, size = 12, weight = 'semibold')
    return ax
```

```python
df_reff = get_SCAM_reff()
```

```python
def show_reff_limits(state = 'liq', logy = True):
    matplotlib.rcParams.update({'xtick.direction': 'out',
                               'ytick.direction': 'out'})
    
    fig, axes = plt.subplots(nrows=1, ncols=2,
                             figsize = (15, 7),
                            dpi = 300)
    if state == 'liq':
        roots_state = [roots_awb, roots_aww]
        derivs_state = [derivs_root_awb, derivs_root_aww]
    elif state in ['ice', 'snw']:
        roots_state = [roots_aib, roots_aiw]
        derivs_state = [derivs_root_aib, derivs_root_aiw]
    else:
        raise InputError('state needs to be liq, ice, or snw')
        
    optical_property_names = ['${}_{{{}}}$'.format(sympy.latex(syc.beta), state), 
                              '${}_{{{}}}$'.format(sympy.latex(syc.omega), state)]
        
    for ax, roots, derivs, name in zip(axes, roots_state, derivs_state,
                                      optical_property_names):
        ax = histogram_reff(ax, df_reff, state = state, logy = logy)
        ax = overlay_reff_fitrange(ax, state = state)
        #ax = overlay_max_reffs(ax, roots, derivs, logy = logy)
        
        ax.set_title(name)
        ax.set_axis_bgcolor((.65, .65, .65))
        ax.set_axisbelow(b = True)
        
        [v.set_color((1, 1, 1)) for k, v in ax.spines.items()]
                
        ax.yaxis.grid(b = False) #, which = 'major', color = (1, 1, 1), linestyle = '--')
        ax.yaxis.set_tick_params(which = 'major', length = 6)
        ax.yaxis.set_tick_params(which = 'minor', length = 3)
        
        ax.set_xlabel('$r_{eff}$')
        ax.xaxis.grid(b = False) #, which = 'major', color = (1, 1, 1), linestyle = '--')
        ax.xaxis.set_tick_params(which = 'major', length = 6)
        ax.xaxis.set_tick_params(which = 'minor', length = 3)

   
    return fig, axes
```
