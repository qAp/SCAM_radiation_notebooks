```python
%matplotlib inline
%config InlineBackend.figure_format = 'png'

import numpy as np
import pandas as pd
import matplotlib
import IPython.html.widgets as weejet
from IPython import display
import xray

matplotlib.rcParams.update({'font.size': 16,
                            'xtick.direction': 'out',
                            'ytick.direction': 'out'})
```

There are three cases to compare here:

1. LWrrtmg SWrrtmg
2. LWclirad SWrrtimg (Yi Hsuan min/max)
3. LWclirad SWrrtmg (no scattering)

Below, numbers 1 to 3 will be in some variable names to identify the cases.

#### load results from all cases

```python
# load case 1
ncpath = '../examples/yihsuans_arm95_rrtmg/0209-123711/camrun.cam.h0.0095-07-18-19800.nc'
h1 = xray.open_dataset(ncpath, decode_times = False)
```

```python
# load case 2
ncpath = '../examples/arm95_LWclirad_SWrrtmg/0306-111654_yihsuan_minmax_scattering/camrun.cam.h0.0095-07-18-19800.nc'
h2 = xray.open_dataset(ncpath, decode_times = False)
```

```python
# load case 3
ncpath = '../examples/arm95_LWclirad_SWrrtmg/0306-153757/camrun.cam.h0.0095-07-18-19800.nc'
h3 = xray.open_dataset(ncpath, decode_times = False)
```

#### convert units of lev

\begin{equation}
(\mathrm{hyam} * \mathrm{P0} + \mathrm{hybm} * \mathrm{PS}) / 100 
\end{equation}

gives pressure in mbars.

```python
pressure = 1e-2 * (h1['hyam'] * h1['P0'] + h1['hybm'] * h1['PS'])\
.transpose('time', 'lev', 'lat', 'lon')
pressure.attrs['units'] = 'mbar'
pressure = pressure[{'time': 0, 'lat': 0, 'lon': 0}]
```

#### CLOUD (cloud fraction), QRL (longwave heating rate) and Q (specific humidity) are compared here

```python
interests = ['CLOUD', 'QRL', 'Q']
```

#### take the difference Case 2 - Case 1
Convert lev coordinates to mbars
Convert heating rates to units of deg/day

```python
diff21 = h2[interests].copy(deep = True) - h1[interests].copy(deep = True)\
.copy(deep = True)

diff21.coords['lev'] = ('lev', pressure, {'units': 'mbar'})

for interest in interests:
    for k in ['long_name', 'units']:
        diff21[interest].attrs[k] = h1[interest].attrs[k]
        
diff21['QRL'] *= 86400.
diff21['QRL'].attrs['units'] = 'K/day'
```

#### take the difference Case 3 - Case 1
Convert lev coordinates to mbar

```python
diff31 = h3[interests].copy(deep = True) - h1[interests].copy(deep = True)
diff31.coords['lev'] = ('lev', pressure, {'units': 'mbar'})

for interest in interests:
    for k in ['long_name', 'units']:
        diff31[interest].attrs[k] = h1[interest].attrs[k]
        
diff31['QRL'] *= 86400.
diff31['QRL'].attrs['units'] = 'K/day'
```

#### take the difference Case 2 - Case 3
Convert lev coordinates to mbar

```python
diff23 = h2[interests].copy(deep = True) - h3[interests].copy(deep = True)
diff23.coords['lev'] = ('lev', pressure, {'units': 'mbar'})

for interest in interests:
    for k in ['long_name', 'units']:
        diff23[interest].attrs[k] = h1[interest].attrs[k]
        
diff23['QRL'] *= 86400.
diff23['QRL'].attrs['units'] = 'K/day'
```

#### label all cases and store them in a dictionary

```python
diffcases = {'CLIRAD scat. - RRTMG': diff21, 
             'CLIRAD no scat. - RRTMG': diff31, 
             'CLIDAD scat. - CLIDAD no scat.': diff23}
```

#### define the plot function for contourf plot of time vs pressure vs  

```python
def plot_diff(axis, datasetname, varname, 
              lev_min = -1., lev_max = 1., lev_step = .1,
              extend = 'neither',
              colorbar_pow = None):
    
    da = datasetname[varname]
    
    x = np.arange(1, da.coords['time'].shape[0] + 1)
    y = da.coords['lev'].values
    Z = da[{'lon': 0, 'lat': 0}].transpose('lev', 'time').values

    cs = axis.contourf(x, y, Z,
                     cmap = matplotlib.cm.coolwarm,
                     levels = np.arange(lev_min, lev_max + lev_step, lev_step),
                     extend = extend)

    if colorbar_pow != None:
        def colorbar_fmt(value, pos):
            return r'{:.2f}'.format(10**(- colorbar_pow) * value)
        cbar = plt.colorbar(cs, 
                            format = matplotlib.ticker.FuncFormatter(colorbar_fmt),
                            ax = axis)
        cbar.set_label(r'$\times 10^{{{}}}$ [{}]'.format(colorbar_pow, da.attrs['units']))
    else:
        cbar = plt.colorbar(cs, ax = axis)
        cbar.set_label('[{}]'.format(da.attrs['units']))   
    
    axis.set_title(da.attrs['long_name'])
    [v.set_color((1, 1, 1)) for k, v in axis.spines.items()]

    axis.set_ylabel('lev [{}]'.format(da.coords['lev'].attrs['units']))
    axis.invert_yaxis()
    axis.yaxis.grid(which = 'major')
    axis.yaxis.set_minor_locator(matplotlib.ticker.MultipleLocator(50))
    axis.yaxis.set_tick_params(which = 'minor', length = 3)
    axis.yaxis.set_tick_params(which = 'major', length = 6)

    axis.set_xlabel('time [index]')
    axis.xaxis.grid(which = 'major')
    axis.xaxis.set_minor_locator(matplotlib.ticker.MultipleLocator(50))
    axis.xaxis.set_tick_params(which = 'minor', length = 3)
    axis.xaxis.set_tick_params(which = 'major', length = 6)
    return axis
```

#### Plot and display all cases for CLOUD, QRL and Q

```python
def plot_diff_allcases(varname, 
                       lev_min = -1, lev_max = 1, lev_step = .1, 
                       extend = 'neither',
                       colorbar_pow = None):
    fig, axes = plt.subplots(nrows = 1, ncols = 3, figsize = (24, 7), dpi = 300)
    for axis, casename in zip(axes, diffcases.keys()):
        axis = plot_diff(axis, diffcases[casename], varname,
                  lev_min = lev_min, lev_max = lev_max, lev_step = lev_step,
                  extend = extend, colorbar_pow = colorbar_pow)
        axis.set_title(casename + '\n' + axis.get_title())
        plt.tight_layout()
```

```python
plot_diff_allcases('CLOUD', lev_min = -1, lev_max = 1, lev_step = .1)
```

```python
plot_diff_allcases('QRL', lev_min = -5, lev_max = 15, lev_step = 5e-1, extend = 'both')
```

```python
plot_diff_allcases('Q', lev_min = - 5e-3, lev_max = + 5e-3, 
          lev_step = 5e-4, extend = 'both',
          colorbar_pow = -2)
```
