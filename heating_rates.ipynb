```python
%matplotlib inline
import os
import pandas as pd
import numpy as np
import netCDF4

import matplotlib.cm as matcm
import matplotlib.pyplot as plt

import IPython.html.widgets as weejet
```

```python
filepath_rrtmg = '../examples/yihsuans_arm95_rrtmg/0209-123711/camrun.cam.h0.0095-07-18-19800.nc'
filepath_clirad = '../examples/yihsuans_arm95_clirad/0210-133728/camrun.cam.h0.0095-07-18-19800.nc'
```

```python
# load rrtmg results
h0rrtmg = netCDF4.Dataset(filepath_rrtmg, 'r')

time_rrtmg = h0rrtmg.variables['time']
lev_rrtmg  = h0rrtmg.variables['lev']
qrl_rrtmg = h0rrtmg.variables['QRL']
qrs_rrtmg = h0rrtmg.variables['QRS']
```

```python
# load clirad results
h0clirad = netCDF4.Dataset(filepath_clirad, 'r')

time_clirad = h0clirad.variables['time']
lev_clirad  = h0clirad.variables['lev']
qrl_clirad = h0clirad.variables['QRL']
qrs_clirad = h0clirad.variables['QRS']
```

```python
# compute difference between the two: CLIRAD - RRTMG
qrls_rrtmg = [qrl_rrtmg, qrs_rrtmg]
qrls_clirad = [qrl_clirad, qrs_clirad]

qrls_rrtmg = [86400 * qr[:, :, 0, 0].T for qr in qrls_rrtmg]
qrls_clirad = [86400 * qr[:, :, 0, 0].T for qr in qrls_clirad]

dqrls = [qr_clirad - qr_rrtmg for qr_clirad, qr_rrtmg in zip(qrls_clirad, qrls_rrtmg)]
```

## Time vs Level vs Heating Rate 

```python
# define function for filled contour plot of some f(x, y) 
def contourf_2D(x, y, Z, axis,
                 xunits = '', yunits = '', zunits = '',
                 xname = '', yname = '', zname = '',
                 contourf_min = -5, contourf_max = 5., contourf_step = .5,
                 invert_yaxis = True):
    '''
    Plot Z = f(x, y) using filled contour. Z.shape = len(y) * len(x) 
    INPUT:
    Z --- 2D array 
    x --- 1D array
    y --- 1D array
    axis --- matplotlib Axes object on which to plot
    OUTPUT:
    axis --- matplotlib Axes object on which to plot
    '''
    X, Y = np.meshgrid(x, y)
    
    contourf_levels = np.arange(contourf_min, contourf_max, contourf_step)
    cs = axis.contourf(X, Y, Z,
                       levels = contourf_levels,  # colour map
                       cmap = matcm.coolwarm, extend = 'both') # colour map
    
    # colour map 
    #cs.cmap.set_under(color = 'magenta')
    #cs.cmap.set_over(color = 'black')
    
    # colour bar
    cbar = plt.colorbar(cs, ax = axis)
    cbar.ax.set_ylabel('{}'.format(zunits))
    
    axis.set_title('{}'.format(zname))
    axis.set_xlabel('{} [{}]'.format(xname, xunits))
    axis.set_ylabel('{} [{}]'.format(yname, yunits))
    
    if invert_yaxis:
        axis.invert_yaxis()
        
    axis.grid(b = True)
    axis.locator_params(axis = 'y', nbins = 10)
    
    #Changing the number of colobar ticks (this has nothing to do with\
    # the number of colours on the colorbar)
    #tick_locator = MaxNLocator(nbins = 6)
    #cb.locator = tick_locator
    #cb.update_ticks()
    return axis
```

```python
def plot_time_level_qr(time, level, qr, axis,
                       contourf_min = -5, contourf_max = 5., contourf_step = .5):
    '''
    Plot heating rate.
    '''
    qr_perday = 86400 * qr[:, :, 0, 0].T
    qr_units = '${}^{\circ} day^{-1}$'
    ax = contourf_2D(time[:], level[:], qr_perday, axis,
                 xname = time.long_name, xunits = time.units, 
                 yname = level.long_name, yunits = level.units,
                 zname = qr.long_name, zunits = qr_units,
                 invert_yaxis = True,
                 contourf_min = contourf_min, contourf_max = contourf_max, contourf_step = contourf_step)
    return ax
```

#### RRTMG

```python
# RRTMG results' heating rates
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (17, 6))
for axis, qr in zip(axes, [qrl_rrtmg, qrs_rrtmg]):
    contourf_min = np.percentile(86400 * qr[:], 2)
    contourf_max = np.percentile(86400 * qr[:], 98)
    
    plot_time_level_qr(time_rrtmg, lev_rrtmg, qr, axis = axis,
                       contourf_min = contourf_min, 
                       contourf_max = contourf_max)
```

#### CLIRAD

```python
# CLIRAD results' heating rates
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (17, 6))
for axis, qr in zip(axes, [qrl_clirad, qrs_clirad]):
    contourf_min = np.percentile(86400 * qr[:], 2)
    contourf_max = np.percentile(86400 * qr[:], 98)
    
    plot_time_level_qr(time_clirad, lev_clirad, qr, axis = axis,
                       contourf_min = contourf_min, 
                       contourf_max = contourf_max)
```

#### CLIRAD - RRTMG

```python
# plot heating rate difference between RRTMG and CLIRAD

znames = [qrl_rrtmg.long_name, qrs_rrtmg.long_name]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (17, 6))
for axis, qr, zname in zip(axes, dqrls, znames):
    contourf_min = np.percentile(qr, 1)
    contourf_max = np.percentile(qr, 99)
    
    contourf_2D(time_rrtmg[:], lev_rrtmg[:], qr,
                axis = axis, 
                zname = zname, zunits = '${}^{\circ} day^{-1}$',
                xname = time_rrtmg.long_name, xunits = time_rrtmg.units,
                yname = lev_rrtmg.long_name, yunits = lev_rrtmg.units,
                contourf_min = contourf_min, 
                contourf_max = contourf_max)
```

## Heating Rate vs Level

```python
def plot_qr_level(lev, qr, i_t = 0):
    '''
    Plot heating rate, qr, versus atmosphere level, lev.
    Qr and lev are on the x and y axes, respectively.
    '''
    fig = plt.figure()
    ax = fig.add_subplot(111,
                         title = 'time {:6.3f} [{}]'.format(time[i_t], 
                                                       time.units),
                         xlabel = 'heating rate [${}^{\circ} day^{-1}$]',
                         ylabel = 'lev [{}]'.format(lev.units))
    
    ax.plot(86400 * qr[i_t, :, 0, 0], lev[:], 
            color = (.7, .5, .1), linewidth = 1.5)
    
    ax.set_xlim(left = 86400 * np.min(qr[:]), 
                right = 86400 * np.max(qr[:]))
    
    
    plt.locator_params(axis = 'y', nbins = 10)
    plt.gca().invert_yaxis()
    plt.grid(b = True)
    return fig
```
