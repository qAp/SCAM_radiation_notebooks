## ARM95 | LW RRTMG | SW RRTMG

```python
%matplotlib inline
import numpy as np
import pandas as pd
import xray
import matplotlib.pyplot as plt
import matplotlib
```

```python
filepath_rrtmg = '../examples/yihsuans_arm95_rrtmg/0209-123711/camrun.cam.h0.0095-07-18-19800.nc'
```

```python
h0 = xray.open_dataset(filepath_rrtmg, decode_times = False)
```

#### convert units of lev

\begin{equation}
(\mathrm{hyam} * \mathrm{P0} + \mathrm{hybm} * \mathrm{PS}) / 100 
\end{equation}

gives pressure in mbars.

```python
pressure = 1e-2 * (h0['hyam'] * h0['P0'] + h0['hybm'] * h0['PS'])\
.transpose('time', 'lev', 'lat', 'lon')
pressure.attrs['units'] = 'mbar'
pressure = pressure[{'time': 0, 'lat': 0, 'lon': 0}]
```

#### convert time to datetime objects

```python
t0 = pd.to_datetime(h0['time'].units.split('since')[-1]) 
timestamps = [t0 + pd.DateOffset(days = dayssince) \
              for dayssince in h0['time'].values]
datetimes = np.array([stamp.to_datetime() for stamp in timestamps])
```

#### make a copy of CLOUD (cloud fraction) and change its lev and time coordindate to the new ones

```python
cloud = h0['CLOUD'].copy()
cloud.coords['time'] = ('time', datetimes)
cloud.coords['lev'] = ('lev', pressure)
```

#### contourf-plot time vs pressure vs CLOUD

```python
x = datetimes
y = cloud['lev'].values
Z = cloud[{'lon': 0, 'lat': 0}].values.T

fig = plt.figure(figsize = (7, 6))
ax = fig.add_subplot(111)

cs = ax.contourf(x, y, Z,
                 cmap = matplotlib.cm.coolwarm,
                 levels = np.arange(Z.min(), Z.max(), .05))

cs.cmap.set_over('yellow')
cs.cmap.set_under('black')

cbar = plt.colorbar(cs)
cbar.set_label('cloud fraction')

# x-axis
ax.set_xlabel('time [datetime]')
ax.xaxis.set_major_locator(matplotlib.dates.MonthLocator())
ax.xaxis.set_major_formatter(matplotlib.dates.DateFormatter('\n%b\n%Y'))
ax.xaxis.set_minor_locator(matplotlib.dates.DayLocator(interval = 2))
ax.xaxis.set_minor_formatter(matplotlib.dates.DateFormatter('%d'))
ax.xaxis.grid(which = 'minor')

# y-axis
ax.set_ylabel('pressure [mbar]')
ax.invert_yaxis()
ax.yaxis.grid(which = 'major')
```
