# Variables ff and ssa_cld that are written out from CLIRAD-LW's cloud_lw() examined

This notebook looks at the calculation of LW scattering due to clouds in CLIRAD.  In the subroutine cloud_lw(), variables *ff*, *ssa_cld*, *tauc*, *tauc_scat*, etc. are written out for each column, layer and waveband for which scattering is calculated.  These values are examined here below.

```python
%matplotlib inline
import numpy as np
import pandas as pd

import matplotlib
import IPython.html.widgets as weejet

matplotlib.rcParams.update({'font.size': 16,
                            'xtick.direction': 'out',
                            'ytick.direction': 'out'})
```

```python
filepath = '/nuwa_cluster/home/jackyu/climate_models/scam/examples/arm95_LWclirad_SWrrtmg/0306-155001/view1.dat'
```

```python
df = pd.read_csv(filepath, 
                 header = 0,
                sep = r'\s+',
                index_col = [0, 1, 2])

df.index.names = ['timestep', 'layer', 'waveband']

df = df.sortlevel(2).sortlevel(1).sortlevel(0)
```

```python
def plot_ff_in_waveband(ibn):
    ibn = int(ibn)
    dfwb = df['ff'].loc[slice(None), slice(None), ibn].unstack(0)
    dfwb = dfwb.reindex(index = range(1, 31 + 1), columns = range(1, 1226 + 1))
    
    x = dfwb.columns.values
    y = dfwb.index.values
    Z = dfwb.values

    lev_min, lev_max, N_lev = 0, 1e3, 40
    
    fig = plt.figure(figsize = (9, 8))
    ax = fig.add_subplot(111)
    cs = ax.contourf(x, y, Z, 
                levels = np.arange(lev_min, lev_max, (lev_max - lev_min) / N_lev),
                cmap = matplotlib.cm.coolwarm,
                extend = 'both')

    cs.cmap.set_over('yellow')
    cs.cmap.set_under('black')

    cbar = plt.colorbar(cs)

    ax.set_title('ff')
    [v.set_color((1, 1, 1)) for k, v in ax.spines.items()]
    ax.set_axis_bgcolor((.6, .6, .6))
    ax.set_axisbelow(b = True)
    
    ax.invert_yaxis()
    
    ax.yaxis.grid(which = 'major', color = 'white', linestyle = '-')
    ax.set_ylabel(dfwb.index.name)
    ax.yaxis.set_minor_locator(matplotlib.ticker.MultipleLocator(1))
    ax.yaxis.set_tick_params(length = 6, which = 'major')
    ax.yaxis.set_tick_params(length = 3, which = 'minor')

    ax.xaxis.grid(which = 'major', color = 'white', linestyle = '-')
    ax.set_xlabel(dfwb.columns.name)
    ax.xaxis.set_minor_locator(matplotlib.ticker.MultipleLocator(50))
    ax.xaxis.set_tick_params(length = 6, which = 'major')
    ax.xaxis.set_tick_params(length = 3, which = 'minor')
```

```python
weejet.interact(plot_ff_in_waveband,
                ibn = [str(i) for i in range(1, 10 + 1)])
```

```python
def plot_ssa_in_waveband(ibn):
    ibn = int(ibn)
    
    dfwb = df['ssa_cld'].loc[slice(None), slice(None), ibn].unstack(0)\
        .reindex(index = range(1, 31 + 1), columns = range(1, 1226 + 1))
        
    x = dfwb.columns.values
    y = dfwb.index.values
    Z = dfwb.values
   
    lev_min, lev_max, N_lev = 0., dfwb.unstack().quantile(q = .95), 40
    
    fig = plt.figure(figsize = (9, 8))
    ax = fig.add_subplot(111)
    cs = ax.contourf(x, y, Z,
                    levels = np.arange(lev_min, lev_max, (lev_max - lev_min) / N_lev),
                    cmap = matplotlib.cm.coolwarm,
                    extend = 'both')
    
    cbar = plt.colorbar(cs)
    
    ax.set_title('ssa_cld')
    ax.set_axisbelow(b = True)
    ax.set_axis_bgcolor((.6, .6, .6))
    [v.set_color((1, 1, 1)) for k, v in ax.spines.items()]
    
    ax.set_ylabel(dfwb.index.name)
    ax.invert_yaxis()
    ax.yaxis.grid(which = 'major', color = (1, 1, 1), linestyle = '-')
    ax.yaxis.set_minor_locator(matplotlib.ticker.MultipleLocator(1))
    ax.yaxis.set_tick_params(length = 6., which = 'major')
    ax.yaxis.set_tick_params(length = 3, which = 'minor')
    
    ax.set_xlabel(dfwb.columns.name)
    ax.xaxis.grid(which = 'major', color = (1, 1, 1), linestyle = '-')
    ax.xaxis.set_minor_locator(matplotlib.ticker.MultipleLocator(50))
    ax.xaxis.set_tick_params(length = 6, which = 'major')
    ax.xaxis.set_tick_params(length = 3, which = 'minor')
```

```python
weejet.interact(plot_ssa_in_waveband,
                ibn = [str(k) for k in range(1, 10 + 1)])
```

```python
def plot_tau_in_waveband(ibn):
    ibn = int(ibn)
    dfwb = (df['tauc'] - df['tauc_scat_b']).loc[slice(None), slice(None), ibn].unstack(0)
    dfwb = dfwb.reindex(index = range(1, 31 + 1), columns = range(1, 1226 + 1))
#    dfwb = dfwb.fillna(0.)
    
    x = dfwb.columns.values
    y = dfwb.index.values
    Z = dfwb.values

    lev_min, lev_max, N_lev = -5, 5, 40
    
    fig = plt.figure(figsize = (9, 8))
    ax = fig.add_subplot(111)
    cs = ax.contourf(x, y, Z, 
                levels = np.arange(lev_min, lev_max, (lev_max - lev_min) / N_lev),
                cmap = matplotlib.cm.coolwarm,
                extend = 'both')

    cs.cmap.set_over('yellow')
    cs.cmap.set_under('black')

    cbar = plt.colorbar(cs)

    ax.set_title('tauc - tauc_scat_b')
    ax.set_axis_bgcolor((.6, .6, .6))
    ax.set_axisbelow(b = True)
    [v.set_color((1, 1, 1)) for k, v in ax.spines.items()]

    ax.invert_yaxis()
    ax.yaxis.grid(which = 'major', color = 'white', linestyle = '-')
    ax.set_ylabel(dfwb.index.name)
    ax.yaxis.set_minor_locator(matplotlib.ticker.MultipleLocator(1))  
    ax.yaxis.set_tick_params(which = 'major', length = 6.)
    ax.yaxis.set_tick_params(which = 'minor', length = 3.)

    ax.xaxis.grid(which = 'major', color = 'white', linestyle = '-')
    ax.set_xlabel(dfwb.columns.name)
    ax.xaxis.set_minor_locator(matplotlib.ticker.MultipleLocator(50))
    ax.xaxis.set_tick_params(which = 'major', length = 6.)
    ax.xaxis.set_tick_params(which = 'minor', length = 3.)
```

```python
weejet.interact(plot_tau_in_waveband, ibn = [str(k) for k in range(1, 10 + 1)])
```
